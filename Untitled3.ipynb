```python
class Dog:
    def __init__(self, name, size, age, breed):
        self.name = name
        self.size = size
        self.age = age
        self.breed = breed

    def bark(self):
        print(f"Woof! I am {self.name}")

    def get_size(self):
        print(f"{self.name} is {self.size} kg")

    def set_name(self, new_name):
        self.name = new_name
        print(f"Name has been changed to {self.name}")

    def set_size(self, new_size):
        self.size = new_size
        print(f"Size has been changed to {self.size} kg")

    def get_breed(self):
        print(f"{self.name} is a {self.breed}")

    def dog_years(self):
        dog_years = self.age * 7
        print(f"{self.name} is {dog_years} dog years old")

dog1 = Dog("Basalt", 50, 4, "Great dane")
dog2 = Dog("Happy", 40, 1, "Golden retriever")
dog3 = Dog("Alex", 45, 9, "Pitbull")

dog1.bark()
dog1.get_size()
dog1.get_breed()
dog1.dog_years()

print()

dog2.bark()
dog2.get_size()
dog2.set_size(27)
dog2.get_breed()
dog2.dog_years()

print()

dog3.bark()
dog3.get_size()
dog3.set_size(12)
dog3.get_breed()
dog3.dog_years()
```

```python
class Book:
    def __init__(self, title, author, price, quantity):
        self.title = title
        self.author = author
        self.price = price
        self.quantity = quantity

    def get_price(self):
        return self.price

    def set_price(self, new_price):
        self.price = new_price
        print(f"The price of '{self.title}' has been updated to {self.price}")

    def get_quantity(self):
        return self.quantity

    def set_quantity(self, new_quantity):
        self.quantity = new_quantity
        print(f"The quantity of '{self.title}' has been updated to {self.quantity}")

    def sell(self, number_sold):
        if number_sold > self.quantity:
            print(f"Not enough stock of '{self.title}' to sell {number_sold} copies.")
        else:
            self.quantity -= number_sold
            print(f"Sold {number_sold} copies of '{self.title}'. {self.quantity} copies remaining.")

    def restock(self, number_added):
        self.quantity += number_added
        print(f"Restocked {number_added} copies of '{self.title}'. {self.quantity} copies now in stock.")

book1 = Book("1984", "George Orwell", 15.99, 10)
book2 = Book("To Kill a Mockingbird", "Harper Lee", 12.99, 5)
book3 = Book("The Great Gatsby", "F. Scott Fitzgerald", 10.99, 8)

print(f"Title: {book1.title}, Author: {book1.author}, Price: ${book1.get_price()}, Quantity: {book1.get_quantity()} copies")
book1.sell(3)
book1.restock(5)
book1.set_price(14.99)
book1.set_quantity(12)
print()

print(f"Title: {book2.title}, Author: {book2.author}, Price: ${book2.get_price()}, Quantity: {book2.get_quantity()} copies")
book2.sell(6) 
book2.restock(10)
book2.set_price(11.99)
book2.set_quantity(15)
print()

print(f"Title: {book3.title}, Author: {book3.author}, Price: ${book3.get_price()}, Quantity: {book3.get_quantity()} copies")
book3.sell(2)
book3.restock(3)
book3.set_price(9.99)
book3.set_quantity(9)
```
